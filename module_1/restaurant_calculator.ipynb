Restoran Hesabı: Yemek fiyatı, bahşiş oranı ve kişi sayısını alıp kişi başı tutarı hesaplayan program      
Restaurant Calculator: A program that calculates the amount per person based on the meal price, tip rate, and number of people.

```python
yemek_fiyati = float(input("Toplam yemek fiyatı: "))
bahsis_orani = float(input("Bahşiş oranı (%) girin: "))
kisi_sayisi = int(input("Kişi sayısı: "))

bahsis_miktari = yemek_fiyati * (bahsis_orani / 100)
toplam_hesap = yemek_fiyati + bahsis_miktari
kisi_basi = toplam_hesap / kisi_sayisi

print(f"\nBahşiş miktarı: {bahsis_miktari:.2f} TL")
print(f"Toplam hesap: {toplam_hesap:.2f} TL")
print(f"Kişi başı ödenecek tutar: {kisi_basi:.2f} TL")
```
